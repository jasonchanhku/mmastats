## Setup Configuration

```python
from bs4 import BeautifulSoup
from itertools import islice
import networkx as nx
import urllib3
import numpy as np
import pandas as pd
import csv
import copy
from py2neo import Graph, Path
import matplotlib.pyplot as plt
import pygraphviz
import neonx


alpha = list(map(chr, range(97, 123)))
all_fighters_dict = {}
all_fighters_list = []

ranking_url = "http://www.fightmetric.com/rankings"

http = urllib3.PoolManager()
ranks = http.request('GET', ranking_url)
fight_soup = BeautifulSoup(ranks.data)
```

## Create map of all fighters

The below creates an initial scrape, looking for all fighters and their associated fighter urls (these are unique and are links to each fighter's personal statistics.)

```python
with open('UFC.csv', 'wb') as csvfile:
    for char in alpha:
        all_fighters_url = "http://www.fightmetric.com/statistics/fighters?char=" + char + "&page=all"
        all_fighters = http.request('GET', all_fighters_url)
        fighter_soup = BeautifulSoup(all_fighters.data)
        fighters_row = fighter_soup.find("tbody")

        for tr in fighters_row.findAll("tr"):
            name = ""
            link = ""
            fullname = ""
            for text in tr.findAll("a"):
                name = name + " " + text.getText()
                link = text['href']
                fullname = " ".join(name.split()[:2])
            all_fighters_dict[fullname] = link
            all_fighters_list.append(fullname)
            if fullname is not None:
                csvwriter = csv.writer(csvfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                csvwriter.writerow([fullname, link])
```

## Individual Fighter Statistics

OK, first we're going to create dataset with all fighters we can, and their individual stats. We'll use this data frame for some rudamentary analysis.

### Run this once - After we have the raw data, we can use the csv directly

First we create the dataframe, from the raw url map we created previously. We then save it to disk so that this doesn't need to be done every time.

```python
all_fighters_df = pd.read_csv('UFC.csv')

# Clean up of csv
all_fighters_df = all_fighters_df.drop('1', 1)
all_fighters_df.columns = ['name', 'url']
all_fighters_df['wins'] = np.nan
all_fighters_df['losses'] = np.nan
all_fighters_df['draws'] = np.nan
all_fighters_df['height'] = np.nan
all_fighters_df['weight'] = np.nan
all_fighters_df['division'] = np.nan
all_fighters_df['reach'] = np.nan
all_fighters_df['stance'] = np.nan
all_fighters_df['dob'] = np.nan
all_fighters_df['SLpM'] = np.nan
all_fighters_df['StrAcc'] = np.nan
all_fighters_df['SApM'] = np.nan
all_fighters_df['StrDef'] = np.nan
all_fighters_df['TDAvg'] = np.nan
all_fighters_df['TDAcc'] = np.nan
all_fighters_df['TDDef'] = np.nan
all_fighters_df['SubAvg'] = np.nan

def get_fighter_stats(url):
    r = http.request('GET', url)
    soup = BeautifulSoup(r.data)
    fighter_stats = []
    fight_stats = []
    # Fighter Record
    record = soup.find("span", class_="b-content__title-record").getText()
    record = record.strip().split()[1].split("-")
    row_index = all_fighters_df.url == url
    all_fighters_df.ix[row_index, ['wins']] = record[0]
    all_fighters_df.ix[row_index, ['losses']] = record[1]
    all_fighters_df.ix[row_index, ['draws']] = record[2]
    
    #Fighter Stats
    div = soup.find("div", class_="b-list__info-box b-list__info-box_style_small-width js-guide").findAll("li")
    for li in div:
        stats = li.getText().strip().split(":")
        fighter_stats.append(stats[1].strip())
    all_fighters_df.ix[row_index, ['height']] = fighter_stats[0]
    all_fighters_df.ix[row_index, ['weight']] = fighter_stats[1]
    all_fighters_df.ix[row_index, ['reach']] = fighter_stats[2]
    all_fighters_df.ix[row_index, ['stance']] = fighter_stats[3]
    all_fighters_df.ix[row_index, ['dob']] = fighter_stats[4]
    
    
    # Fight Stats
    div = soup.find("div", class_="b-list__info-box-left clearfix").findAll("li")
    for li in div:
        stats = li.getText().strip().split(":")
        if stats[0] != "":
            fight_stats.append(stats[1].strip())
    all_fighters_df.ix[row_index, ['SLpM']] = fight_stats[0]
    all_fighters_df.ix[row_index, ['StrAcc']] = fight_stats[1]
    all_fighters_df.ix[row_index, ['SApM']] = fight_stats[2]
    all_fighters_df.ix[row_index, ['StrDef']] = fight_stats[3]
    all_fighters_df.ix[row_index, ['TDAvg']] = fight_stats[4]
    all_fighters_df.ix[row_index, ['TDAcc']] = fight_stats[5]
    all_fighters_df.ix[row_index, ['TDDef']] = fight_stats[6]
    all_fighters_df.ix[row_index, ['SubAvg']] = fight_stats[7]
    
    if fighter_stats[1] is not None and fighter_stats[1].split(' ')[0] != "--":
        w = fighter_stats[1].split(' ')[0]
        w = int(w)
        if w < 115:
            all_fighters_df.ix[row_index, ['division']] = "Strawweight"
        if w >= 115 and w < 126:
            all_fighters_df.ix[row_index, ['division']] = "Flyweight"
        if w >=126 and w <= 135:
            all_fighters_df.ix[row_index, ['division']] = "Bantamweight"
        if w >= 136 and w <= 145:
            all_fighters_df.ix[row_index, ['division']] = "Featherweight"
        if w >= 146 and w <= 155:
            all_fighters_df.ix[row_index, ['division']] = "Lightweight"
        if w >= 156 and w <= 170: 
            all_fighters_df.ix[row_index, ['division']] = "Welterweight"
        if w >= 171 and w <= 185:
            all_fighters_df.ix[row_index, ['division']] = "Middleweight"
        if w >= 186 and w <= 205:
            all_fighters_df.ix[row_index, ['division']] = "Light Heavyweight"
        if w >= 206 and w <= 265:
            all_fighters_df.ix[row_index, ['division']] = "Heavyweight"
        if w > 265:
            all_fighters_df.ix[row_index, ['division']] = "Super Heavyweight"
```

```python
all_fighters_df.url.apply(lambda url: get_fighter_stats(url))
all_fighters_df.to_csv('UFC_DF.csv', sep=",")
```

## Wins, losses, and draws

The below simply adds some useful information about wins, losses and draws to our dataframe.

```python
all_fighters_df['wins'][:] = all_fighters_df.wins.map(int)
all_fighters_df['losses'][:] = all_fighters_df.losses.map(int)
all_fighters_df['draws'][:] = all_fighters_df.draws.map(int)
all_fighters_df['total'] = all_fighters_df[['wins','losses','draws']].sum()
all_fighters_df['wins-norm'] = np.nan
all_fighters_df['losses-norm'] = np.nan
all_fighters_df['draws-norm'] = np.nan
all_fighters_df['total'] = all_fighters_df[['wins','losses','draws']].sum(axis=1)
all_fighters_df['wins-norm'] = all_fighters_df.wins / all_fighters_df.total
all_fighters_df['losses-norm'] = all_fighters_df.losses / all_fighters_df.total
all_fighters_df['draws-norm'] = all_fighters_df.draws / all_fighters_df.total
```

From here we look at the social network of fighters, to get a better understanding of who fought who.
